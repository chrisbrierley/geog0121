## Welcome to the Jupyter Notebook system
Jupyter Notebooks are a way to create documents that have both interactive (REPL) code and documentation!  Jupyter works with Python, Julia, R, and many other languages. What this means is that we combine programming code along with its explanations and results. Where possible I will be using these Notebooks during the practical sessions.

Each cell in a notebook can be Markdown or code.
 - [Enter] Puts you in cell-editing mode, and [Escape] brings you to document-editing mode.
 - [Shift-Enter] Evaluates the cell.
 - [M] Switches the cell to Markdown mode, and [Y] switches it to code.
 - [H] Shows you help with other commands.
